```python
# Subsetting the data to the desired columns
cols = ['STATE','zipcode','N1','A00100','A02650']
df_clean = df[cols]
df_clean.head()
```

```python
df_clean.info()
```

```python
# rename columns
new_cols = {'STATE':'st','zipcode':'zip','N1':'n_returns','A00100':'agi','A02650':'total_income'}
df_clean.rename(new_cols, axis=1, inplace=True)
df_clean.head()
```

It appears there are 0 values in the ZIPCODE column. Let's see how many of those there are.

```python
# count total number of rows where ZIPCODE is 0
len(df_clean[df_clean['zip'] == 0])
```

Let's replace these zeros with NaN

```python
# replace zeros with NaN
df_clean.zip.replace(0, np.nan, inplace=True)
df_clean.isna().sum()
```

Let's take a look at the top 5 observations of a few of the variables.

By Number of Returns

```python
# sort by n_returns
df_clean.sort_values('n_returns', ascending=False).head()
```

Odd that the areas with the largest number of returns do not specify ZIPCODE. 

By AGI

```python
# sort by agi
df_clean.sort_values('agi', ascending=False).head()
```

By Total Income

```python
# sort by total_income
df_clean.sort_values('total_income', ascending=False).head()
```

```python
# calculate diff between total income and agi
df_clean['balance'] = df_clean.total_income - df_clean.agi
df_clean.head()
```

```python
# convert all states to lower case
df_clean.st = df_clean.st.str.lower()
df_clean.head()
```

```python
# convert them back!
df_clean.st = df_clean.st.str.upper()
df_clean.head()
```

```python
# group by st
df_clean.groupby('st')['n_returns'].mean().sort_values(ascending=False)
```

```python
# get the min
# group by st
df_clean.groupby('st')['n_returns'].min().sort_values(ascending=False)
```

```python
# group by st
df_clean.groupby('st')['n_returns'].max().sort_values(ascending=False)
```

```python
# by two variables
df_clean.groupby(['st','zip'])['total_income'].max()
```
