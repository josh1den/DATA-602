##### Load your dataset into python using the pandas library and save data into a dataframe named dfi (where i is one of your datasets, for a total of 4).  

```python
# importing datasets
# weather URL from GitHub (source: https://github.com/data602sps/datasetspractice)
weather_url = "https://raw.githubusercontent.com/data602sps/datasetspractice/main/weather.csv"
# df1 : weather data
df1 = pd.read_csv(weather_url)

# salary URL from GitHub (source: https://github.com/data602sps/datasetspractice)
salary_url = "https://raw.githubusercontent.com/data602sps/datasetspractice/main/salarydata.csv"
# df2 : salary data
df2 = pd.read_csv(salary_url)

# birds URL from data.gov (source: https://catalog.data.gov/dataset/riparian-birds-sierra-nevada-foothills-ds303-3e51a)
birds_url = "https://data-cdfw.opendata.arcgis.com/datasets/e1b1ff8f4872409eb43447175105e8ce_0.csv?outSR=%7B%22latestWkid%22%3A3857%2C%22wkid%22%3A102100%7D"
# df3 : birds data
df3 = pd.read_csv(birds_url)

# tate artist data from GitHub (source: https://github.com/tategallery/collection/blob/master/artist_data.csv)
tate_url = "https://raw.githubusercontent.com/tategallery/collection/master/artist_data.csv"
# df4 : tate data
df4 = pd.read_csv(tate_url)
```
